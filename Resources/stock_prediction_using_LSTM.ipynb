### Trying Different Model Variants
To improve my model, I manually experimented with different LSTM configurations.
I changed:
- number of LSTM units

Below are the different attempts I tried and I compared their accuracy/loss.

# Predict next n future days stock price

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from tensorflow.keras.models import load_model
import pickle


# Predict next n future days

def predict_future_days(model, testing_data, scaler, n_days, look_back=100):

    raw = testing_data.values if hasattr(testing_data, "values") else testing_data

    last_100_raw = raw[-look_back:]
    last_100_scaled = scaler.transform(last_100_raw)

    current_input = last_100_scaled.reshape(1, look_back, 1)

    preds_scaled = []

    for _ in range(n_days):
        next_scaled = model.predict(current_input)[0][0]
        preds_scaled.append(next_scaled)

        current_input = np.append(
            current_input[:, 1:, :],
            [[[next_scaled]]],
            axis=1
        )

    preds_scaled = np.array(preds_scaled).reshape(-1, 1)
    preds = scaler.inverse_transform(preds_scaled)

    return preds



# USER INPUT
n = int(input("Enter number of future days to predict: "))

scaler = MinMaxScaler(feature_range=(0,1))
model = load_model("stock_prediction_model.keras")
data_testing = pd.read_csv("stock_data.csv")

future_prices = predict_future_days(model, data_testing, scaler, n_days=n)
future_prices
```
